## 該如何執行一段 Python 程式？

- 在終端機使用 `python` 指令執行副檔名為 `.py` 的程式碼
- 執行一個或多個副檔名為 `.ipynb` Jupyter Notebook 裡頭的 Code Cell(s)

## 根據專案任務決定怎麼執行 Python 程式

- 程式開發：執行 Python 程式碼 `.py`
- 資料分析：執行 Jupyter Notebook `.ipynb`
